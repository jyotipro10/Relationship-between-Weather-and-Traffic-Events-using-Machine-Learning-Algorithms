```
def sigmoid(z) :
    return 1/(1 + np.exp(-z))
```

```
## initialize w as a vector of zeros with size dim and set value of b to 0
def initialize_with_zeros(dim) :
    w = np.zeros((dim,1))
    b = 0
    return w,b
```

```
def propagate(w,b,X,y) :
    m = np.shape(X)[1]
    A = sigmoid(np.dot(w.T,X) + b)
    cost = (-1/m) * np.sum((y*(np.log(A))) + ((1 - y)*(np.log(1 - A))))
    dw = np.dot(X,(A - y).T)/m
    db = np.sum(A - y)/m
    
    #assert(dw.shape == w.shape)
    #assert(db.dtype == float)
    cost = np.squeeze(cost)
    #assert(cost.shape == ())
    
    grads = {'dw' : dw,'db' : db}
    
    return grads,cost
```

```
def optimize(w,b,X,y,num_iterations,learning_rate,print_cost = False) :
    costs = []
    for i in range(num_iterations) :
        grads,cost = propagate(w,b,X,y)
        dw = grads['dw']
        db = grads['db']
        w = w - (learning_rate * dw)
        b = b - (learning_rate * db)
        
        if i % 100 == 0 :
            costs.append(cost)
        if (print_cost) and (i % 100 == 0) :
            print('cost after %i iterations is : %f' %(i,cost))
        
        params = {'w' : w, 'b' : b}
        grads = {'dw' : dw, 'db' : db}
        
    return params,grads,costs
```

```
def predict(w,b,X) :
    y_hat = sigmoid(np.dot(w.T,X) + b)
    #iterations = np.shape(X)[1]
    #y_prediction = np.zeros((1,iterations))
    # for i in range(iterations) :
    #     if y_hat[0,i] > 0.5 :
    #         y_prediction[0,i] = 1
    #     else :
    #         y_prediction[0,i] = 0
    #     pass
    #assert(y_prediction.shape == (1,iterations))
    return y_hat
```

```
def model(X_train,Y_train,X_test,Y_test,num_iterations,learning_rate, print_cost) :
    # initializing w and b with zeros
    w,b = initialize_with_zeros(X_train.shape[0])
    # optimizing w,b to minimize cost. performing gradient descent
    parameters,grads,costs = optimize(w,b,X_train,Y_train,num_iterations,learning_rate,print_cost)
    w = parameters['w']
    b = parameters['b']
    dw = grads['dw']
    db = grads['db']
    # prdictions of train and test sets
    Y_prediction_train = predict(w,b,X_train)
    Y_prediction_test = predict(w,b,X_test)
    
    ## print test and train errors
    print('train accuracy {}%'.format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print('test accuracy {}%'.format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))
    ### dictionary of informations about the model
    d = {
        'costs' : costs,
        'train prediction' : Y_prediction_train,
        'test prediction' : Y_prediction_test,
        'w' : w,
        'b' : b,
        'num_iterations' : num_iterations,
        'learning rate' : learning_rate
    }
    
    return d
```
