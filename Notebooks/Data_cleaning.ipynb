```
l = len(df)
df['lat'] = 0
df['lng'] = 0
df
for i in range(l) :
  row = df.iloc[i]
  if row['EventId'][0] == 'W' :
    df.iloc[i,df.columns.get_loc('lat')] = row['wLat']
    df.iloc[i,df.columns.get_loc('lng')] = row['wLng']
  elif row['EventId'][0] == 'T' :
    df.iloc[i,df.columns.get_loc('lat')] = row['tLat']
    df.iloc[i,df.columns.get_loc('lng')] = row['tLng']
df
```

```
del df['wLat']
del df['wLng']
del df['tLat']
del df['tLng']
df
```

```
df.to_csv('cleanData.csv')
```
