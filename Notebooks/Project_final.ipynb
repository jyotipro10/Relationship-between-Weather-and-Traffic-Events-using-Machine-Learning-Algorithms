```
import geopy.distance
def calcDist(lat1,lon1,lat2,lon2) :

  coords_1 = (lat1, lon1)
  coords_2 = (lat2, lon2)

  return geopy.distance.distance(coords_1, coords_2).km
```

```
sec_in_day = 24*60*60
def diffTime(first_time,second_time) :
  diff = second_time - first_time
  dt.timedelta(0, 8, 562000)
  
  return list(divmod(diff.days * sec_in_day + diff.seconds, 60))
```
